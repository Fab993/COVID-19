```python
# Import Relevant Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
```

```python
# Import and manipulate data
df = pd.read_csv('/Users/Faber/Documents/GitHub/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
df = df.drop(['Province/State', 'Lat', 'Long'], axis=1)
df.head()
```

```python
# Select train and test: test on Belgium
#train    = df.loc[(df['Country/Region'] == "Italy" )         |  
#                  (df['Country/Region'] == 'Netherlands')    |
#                  (df['Country/Region'] == "Spain" )         | 
#                  (df['Country/Region'] == "Germany")        |
#                  (df['Country/Region'] == "United Kingdom") |
#                  (df['Country/Region'] == "Portugal")       |
#                  (df['Country/Region'] == "Austria")        |
#                  (df['Country/Region'] == "France")         ]

train    = df.loc[(df['Country/Region'] == "Italy" )         |  
                  (df['Country/Region'] == 'Netherlands')    |
                  (df['Country/Region'] == "Spain" )         | 
                  (df['Country/Region'] == "Germany")        |
                  (df['Country/Region'] == "Portugal")       |
                  (df['Country/Region'] == "France")         ]
    
test     = df.loc[df["Country/Region"]=="Belgium"]
end      = train.shape[1]
X_train  = train.iloc[:, 1:end-1]                        # Use past values as X
y_train  = train.iloc[:, -1]                             # Use today's value as y
X_test   = test.iloc[:, 1:end-1]                         # Use past value as X
```

```python
# Fit random forest
regressor = RandomForestRegressor(n_estimators=500, random_state=10, verbose=1) 
regressor.fit(X_train, y_train)
y_pred    = regressor.predict(X_test)
```

```python
# Printe estimated vs. actual numbers
print("--------------------------------------")
print("Results of a simple random forest (train on 22/01 - 22/03): \n")
print("Predicted number of cases: ", float(y_pred))
print("Actual number of cases: ", float(test.iloc[0, -1]))
print("--------------------------------------")
```

```python
# How many confirmed cases there will be tomorrow?
X_train   = train.iloc[:, 1:end-1]                        # Use past values as X
y_train   = train.iloc[:, -1]                             # Use today's value as y
X_test    = test.iloc[:, 2:end]                           # Use past value as X (remove very first value of the series)
regressor = RandomForestRegressor(n_estimators=500, random_state=10, verbose=0) # Set verbose = 1 to display convergence message
regressor.fit(X_train, y_train)
y_pred    = regressor.predict(X_test)
```
